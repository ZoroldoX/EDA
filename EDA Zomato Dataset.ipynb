### Countries with max white rating

```python
df_notrated = final_df.Country[final_df['Rating text'] == 'Not rated']
```

```python
df_notrated = df_notrated.reset_index()
```

```python
df_notrated.drop('index',axis = 1)
```

```python
df_notrated_counts = df_notrated.groupby(['Country']).size()
```

```python
df_notrated_counts = df_notrated_counts.reset_index().rename(columns={0:'vc'})
df_notrated_counts
```

```python
country_list = df_notrated_counts.Country.values
country_counts = df_notrated_counts.vc.values
```

```python
plt.pie(country_counts,labels = country_list)
plt.show()
```

### Pie chart for cities distribution

```python
final_df.columns
```

```python
df_bycity = final_df.groupby('City').size().reset_index().rename(columns={0:"vals"})
```

```python
df_bycity.sort_values(by='vals',inplace=True)
```

```python
df_bycity_tail = df_bycity.tail()
```

```python
plt.pie(df_bycity_tail.vals,labels=df_bycity_tail.City,autopct='%1.2f%%')
plt.show()
```

### Top 10 cuisines

```python
cuisines_df= final_df.groupby('Cuisines').size().reset_index().rename(columns={0:'Vals'})
```

```python
cuisines_df.sort_values(by = 'Vals',inplace=True)
```

```python
cuisines_df
```

```python
cuisines_topten = cuisines_df.tail(10)
```

```python
plt.pie(cuisines_topten.Vals,labels=cuisines_topten.Cuisines,autopct = '%1.3f%%')
plt.show()
```
